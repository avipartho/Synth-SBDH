```python
import json
import os
import numpy as np
import pandas as pd
from IPython.display import display,HTML
```

```python
for dataset in [
    # 'sbdh_gpt4_v2', 'sbdh_gpt4_v3','sbdh_gpt4_msf', 'sbdh_gpt4_msf_v3'
    'sbdh_gpt4_v3'
    ]:
    print('#'*20)
    print(dataset)
    print('#'*20)
    # for model in ['t5-v1_1-base','t5-v1_1-base-ptr','flan-t5-base','flan-t5-base-ptr']:
    for model in ['t5-v1_1-base','flan-t5-base']:
        print('*'*10,model,'*'*10)
        result_dict = {}
        result = None
        for task_type in ['standard', 'task_prefix']:
            result_dict[task_type] = {
                # 'loss':[],
                'micro-f':[],
                'macro-f':[],
            }
            if not os.path.isfile(f'./ckpts/{dataset}/{model}/{task_type}/palm/1.0/gt/0.5/256/64/AdamW/5e-05/2/predict_results.json'):continue
            for seed in ['0','1','2']:
                with open(f'./ckpts/{dataset}/{model}/{task_type}/palm/1.0/gt/0.5/256/64/AdamW/5e-05/{seed}/predict_results.json') as f:
                    result = json.load(f)
                    # result_dict[task_type]['loss'] += [result['test_loss']]
                    result_dict[task_type]['macro-f'] += [result['test_macro-f']]
                    result_dict[task_type]['micro-f'] += [result['test_micro-f']]
            for k,v in result_dict[task_type].items():
                result_dict[task_type][k] = f'{np.average(result_dict[task_type][k])*100:.4}±{np.std(result_dict[task_type][k])*100:.2}'
                
            res_df = pd.DataFrame(result_dict)
            try:res_df['delta'] = res_df.apply(
                lambda x: '{:.2f} ({:.2f}%)'.format(
                    float(x['task_prefix'].split('±')[0])-float(x['standard'].split('±')[0]),
                    (float(x['task_prefix'].split('±')[0])-float(x['standard'].split('±')[0]))*100/float(x['standard'].split('±')[0])
                    ), axis=1
                )
            except:pass
        if result is not None:display(res_df.T)
```
